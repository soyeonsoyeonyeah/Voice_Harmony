```python
import os
import json
import numpy as np
import pathlib
from tqdm import tqdm
import shutil
```

```python
keypoint_root_folder_path = "C:/ict_sign_language_project01/data/keypoint"
keypoint_folder_path_list = os.listdir(keypoint_root_folder_path)
keypoint_folder_path_list
```

```python
for keypoint_folder_path in keypoint_folder_path_list[0:2]:
    print("=" * 100)
    print("keypoint_folder_path=", keypoint_folder_path)
    print("=" * 100)

    json_file_folder = keypoint_root_folder_path + "/" + keypoint_folder_path

    #json_file_folder에 저장된 JSON 파일 리스트 리턴
    json_file_list = os.listdir(json_file_folder)
    json_num = len(json_file_list)

    for i in range(1):
        json_file_path = f"{json_file_folder}/{keypoint_folder_path}_{i}_keypoints.json"

        with open(json_file_path, "r") as keypoint_json_file:
            #키포인트가 저장된 JSON 파일의 내용 리턴
            keypoint_json = json.load(keypoint_json_file)
            
            #포즈, 왼손, 오른손의 키포인트 리턴
            pose_keypoints_2d = keypoint_json["people"][0]["pose_keypoints_2d"]
            hand_left_keypoints_2d = keypoint_json["people"][0]["hand_left_keypoints_2d"]
            hand_right_keypoints_2d = keypoint_json["people"][0]["hand_right_keypoints_2d"]

            #포즈의 키포인트가 저장된 리스트를 numpy 배열로 변환
            #리스트를 3열로 변환하고 행은 생략
            pose_keypoint_arr = np.array(pose_keypoints_2d).reshape(-1, 3)
            hand_left_keypoint_arr = np.array(hand_left_keypoints_2d).reshape(-1, 3)
            hand_right_keypoint_arr = np.array(hand_right_keypoints_2d).reshape(-1, 3)
            print("pose_keypoint_arr=", pose_keypoint_arr)
            print("=" * 100)

            #키포인트의 x, y좌표 리턴(모든 행, 2열 미만)
            print("pose_keypoint_arr[: , 0:2]=", pose_keypoint_arr[: , :2])
            
            #포즈, 왼손, 오른손 키포인트의 x, y좌표 배열들을 합쳐 새로운 키포인트 배열 생성
            keypoint_arr = np.concatenate((pose_keypoint_arr[:, :2], hand_left_keypoint_arr[:, :2], hand_right_keypoint_arr[:, :2]))
            print("keypoint_arr=", keypoint_arr)
            print("=" * 100)
            
```

```python
#배열을 저장할 폴더
save_root_path = f"data/keypoint_arr/"

#배열을 저장할 폴더의 정보를 알아낼 객체 리턴
arr_save_path = pathlib.Path(save_root_path)

if arr_save_path.exists():
    shutil.rmtree(save_root_path)

os.makedirs(save_root_path)
```

```python
for keypoint_folder_path in tqdm(keypoint_folder_path_list):
    json_file_folder = keypoint_root_folder_path + "/" + keypoint_folder_path
    json_file_list = os.listdir(json_file_folder)
    json_num = len(json_file_list)
    
    #JSON 파일의 내용을 저장할 리스트
    all_keypoint_list = []
    
    for i in range(json_num):
        json_file_path = f"{json_file_folder}/{keypoint_folder_path}_{i}_keypoints.json"
        
        with open(json_file_path, "r") as keypoint_json_file:
            #키포인트가 저장된 JSON 파일의 내용 리턴
            keypoint_json = json.load(keypoint_json_file)
            #포즈, 왼손, 오른손 키포인트 리턴
            pose_keypoints_2d = keypoint_json["people"][0]["pose_keypoints_2d"]
            hand_left_keypoints_2d = keypoint_json["people"][0]["hand_left_keypoints_2d"]
            hand_right_keypoints_2d = keypoint_json["people"][0]["hand_right_keypoints_2d"]

            #리스트를 numpy 배열로 변환(행은 생략, 3열)
            pose_keypoint_arr = np.array(pose_keypoints_2d).reshape(-1, 3)
            hand_left_keypoint_arr = np.array(hand_left_keypoints_2d).reshape(-1, 3)
            hand_right_keypoint_arr = np.array(hand_right_keypoints_2d).reshape(-1, 3)
            
            #각 배열들의 x, y좌표를 합쳐 새로운 배열 생성
            keypoint_arr = np.concatenate((pose_keypoint_arr[:, :2], hand_left_keypoint_arr[:, :2], hand_right_keypoint_arr[:, :2]))
            
            #keypoint_arr을 all_keypoint_list에 추가
            all_keypoint_list.append(keypoint_arr.tolist())

    #all_keypoint_list를 실수타입 배열로 변환
    all_keypoint_arr = np.array(all_keypoint_list, dtype="float32")

    if keypoint_folder_path[-1]=='F':
        #all_keypoint_arr을 data.keypoint_arr 폴더에 저장
        np.save(f"{save_root_path}/{keypoint_folder_path[:-2]}", all_keypoint_arr)
    else:
        #all_keypoint_arr을 data.keypoint_arr 폴더에 저장
        np.save(f"{save_root_path}/{keypoint_folder_path}", all_keypoint_arr)
```
