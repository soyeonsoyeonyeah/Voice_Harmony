```python
import random
import numpy as np
import pickle
import pandas as pd
import json
import pathlib
```

```python
random.seed(0)
np.random.seed(0)
```

```python
MAX_FRAME_NUM = 240
```

```python
df = pd.read_csv("data/y_train_preprocess.csv")
```

```python
all_frame_list = []

#0~1행 video_id 컬럼 값 리턴(비디오 이름)
for name in df.loc[0:1, "video_id"]:
    try:
        #비디오의 키포인트 배열을 읽어서 리턴
        X_data = np.load(f'data/keypoint_arr/{name}.npy')
        
    
        frame_keypoint_list = []
    
        #X_data의 1행씩 frame_keypoint에 대입
        for frame_keypoint in X_data:
            x = frame_keypoint[:, 0]
            y = frame_keypoint[:, 1]
    
            #키포인트의 x, y 좌표 평균
            mean_x = np.mean(x)
            mean_y = np.mean(y)
    
            #키포인트의 x, y좌표 표준편차
            std_x = np.std(x)
            std_y = np.std(y)
    
            #키포인트의 x, y 좌표에서 평균을 빼고 표준편차로 나눔
            normal_x = (x - mean_x) / std_x
            normal_y = (y - mean_y) / std_y
    
            #0번째 열에 normal_x, 1번째 열에 normal_y 대입
            frame_keypoint[:, 0] = normal_x
            frame_keypoint[:, 1] = normal_y
            print("frame_keypoint=", frame_keypoint[0:3])
    
            #frame_keypoint(2차원배열)을 1차원배열로 변환한 뒤 리스트로 변환
            frame_keypoint_list.append(frame_keypoint.flatten().tolist())
    
        print("len(frame_keypoint_list)=", len(frame_keypoint_list))
    
        if len(frame_keypoint_list) < MAX_FRAME_NUM:
            random_choice_frame = np.random.choice(len(frame_keypoint_list), MAX_FRAME_NUM - len(frame_keypoint_list))
            random_choice_frame.sort()
            print("random_choice_frame=", random_choice_frame)
    
            #random_choice_frame에 저장된 난수를 random_idx에 대입
            for random_idx in random_choice_frame:
                #frame_keypoint에서 random_idx번째 프레임을 리스트에 추가
                #240행(프레임 키포인트 x, y좌표)으로 길이를 맞춤
                frame_keypoint_list.append(frame_keypoint_list[random_idx])
        else:
            #MAX_FRAME_NUM행(프레임 키포인트 x, y좌표) 리턴
            frame_keypoint_list = frame_keypoint_list[: MAX_FRAME_NUM]
    
        #frame_keypoint의 행을 역순으로 변환
        frame_keypoint_list = frame_keypoint_list[::-1]
    
        all_frame_list.append(frame_keypoint_list)
        
    except Exception as e:    
        print('예외가 발생했습니다.', e)
        
```

```python
all_frame_list = []

#모든행 video_id 컬럼 값 리턴(비디오 이름)
for name in df.loc[: , "video_id"]:
    try:
        #비디오의 키포인트 배열을 읽어서 리턴
        X_data = np.load(f'data/keypoint_arr/{name}.npy')
    
        frame_keypoint_list = []
    
        for frame_keypoint in X_data:
            x = frame_keypoint[:, 0]
            y = frame_keypoint[:, 1]
    
            #키포인트의 x, y 좌표 평균
            mean_x = np.mean(x)
            mean_y = np.mean(y)
    
            #키포인트의 x, y좌표 표준편차
            std_x = np.std(x)
            std_y = np.std(y)
    
            #키포인트의 x, y 좌표에서 평균을 빼고 표준편차로 나눔
            normal_x = (x - mean_x) / std_x
            normal_y = (y - mean_y) / std_y
    
            #0번째 열에 normal_x, 1번째 열에 normal_y 대입
            frame_keypoint[:, 0] = normal_x
            frame_keypoint[:, 1] = normal_y
    
            #frame_keypoint(2차원배열)을 1차원배열로 변환한 뒤 리스트로 변환
            frame_keypoint_list.append(frame_keypoint.flatten().tolist())
    
        if len(frame_keypoint_list) < MAX_FRAME_NUM:
            random_choice_frame = np.random.choice(len(frame_keypoint_list), MAX_FRAME_NUM - len(frame_keypoint_list))
            random_choice_frame.sort()
    
            #random_choice_frame에 저장된 난수를 random_idx에 대입
            for random_idx in random_choice_frame:
                #frame_keypoint에서 random_idx번째 프레임을 리스트에 추가
                #240행(프레임 키포인트 x, y좌표)으로 길이를 맞춤
                frame_keypoint_list.append(frame_keypoint_list[random_idx])
        else:
            #MAX_FRAME_NUM행(프레임 키포인트 x, y좌표) 리턴
            frame_keypoint_list = frame_keypoint_list[: MAX_FRAME_NUM]
    
        #frame_keypoint의 행을 역순으로 변환
        frame_keypoint_list = frame_keypoint_list[::-1]
    
        all_frame_list.append(frame_keypoint_list)
    except Exception as e:    
        print('예외가 발생했습니다.', e)
        
```
