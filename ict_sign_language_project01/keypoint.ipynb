```python
import subprocess
import os
import pathlib
```

```python
#openpose 모델을 저장한 폴더의 경로 설정
os.environ["model_folder"] = "C:/ict_sign_language_project01/openpose/models"
```

```python
#프레임을 저장한 폴더의 서브 폴더 리스트 리턴
frame_path_list = os.listdir("C:/ict_sign_language_project01/data/frame")
```

```python
frame_path_list
```

```python
#키포인트를 저장할 폴더 경로
file_path = "data/keypoint"

#file_path의 정보 리턴
save_path = pathlib.Path(file_path)

if not save_path.exists():
    os.makedirs(file_path)
```

```python
for path in frame_path_list:
    command = f"openpose\\bin\\OpenPoseDemo.exe    --image_dir data\\frame\\{path}    --hand --model_folder openpose\\models --write_json data\\keypoint\\{path}  --display 0 --render_pose 0 --num_gpu 1 --num_gpu_start 0"
    print("=" * 100)
    print(command)
    print("=" * 100)
    subprocess.run(command)
```
